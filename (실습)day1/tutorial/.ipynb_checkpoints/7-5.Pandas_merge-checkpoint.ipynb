# 7.5.Pandas_merge

    여러개의 Dataframe을 merge 함수를 사용하여 병합할 수 있다.
    https://pandas.pydata.org/pandas-docs/stable/merging.html

## DataFrame 병합(merge)

```python
from IPython.display import display
import pandas as pd

left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})
display(left)


right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
display(right)


result = pd.merge(left, right, on='key') # key 컬럼을 기준으로 병합
display(result)
```

## Inner merge : 기준이 되는 컬럼의 동일한 벨류들끼리만 병합함

```python
import pandas as pd

left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})
display(left)

right = pd.DataFrame({'key': ['K2', 'K3', 'K4', 'K5'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
display(right)

result = pd.merge(left, right, how = 'inner', on='key') # key 컬럼을 기준으로 병합
display(result)
```

## outer merge : 기준에 되는 컬럼에 해당되는 값이 없을경우 NaN 값으로 설정하여 모두 합침

```python
import pandas as pd

left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})
display(left)


right = pd.DataFrame({'key': ['K2', 'K3', 'K4', 'K5'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
display(right)


result = pd.merge(left, right, how = 'outer', on='key') # key 컬럼을 기준으로 병합
display(result)
```

## Series 객체 병합

```python
s1 = pd.Series([1,3], index=['a','b'])
s2 = pd.Series([5,6,7], index=['a','b','c'])
s3 = pd.Series([11,13], index=['a','b'])

# 행 단위 결합 
result = pd.concat([s1, s2, s3], axis = 0) # 행 단위 결합 
display(result)
```

```python
# 열 단위 결합 : DF 생성
result2 = pd.concat([s1, s2, s3], axis = 1) # 열 단위(index) 결합 
display(result2)
```
