# 7-4-2.Pandas_descriptive statistics(EDA)

```python
from IPython.display import display
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
```

```python
# 전체 데이터 확인해보기
display(data)
```

```python
# 일부 데이터 확인해보기

display(data.head(10)) # 10줄

display(data.tail())
```

## 데이터 파악하기

```python
# 1. 데이터 shape 확인
display(data.shape)
```

```python
# 2. 데이터 구조 확인
display(data.info())
```

```python
# 3. 데이터 요약통계량 구하기
# 숫자형 자료형태에 대해서만 요약
display(data.describe())
```

```python
# 4. 문자열 변수 분포 확인하기
display(data.groupby('species').size())
```

## 시각화 하기

```python
# 시각화 하기
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.boxplot.html
import matplotlib.pyplot as plt

# 박스플랏
data.boxplot()
plt.show()
```

```python
# 히스토그램
data.hist(edgecolor='black', linewidth=1.2)
plt.show()
```

```python
# scatter_matrix
from pandas.plotting import scatter_matrix

scatter_matrix(data, alpha=1, figsize=(6, 6))
plt.show()
```

```python
# scatter_matrix을 사용할 수 있는 seaborn 모듈 사용
# https://seaborn.pydata.org/generated/seaborn.pairplot.html

import seaborn as sns

sns.pairplot(data, hue="species", size=3)
plt.show()
```
