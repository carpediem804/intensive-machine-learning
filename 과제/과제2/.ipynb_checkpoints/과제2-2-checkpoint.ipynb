```python
import pandas as pd
from pandas import Series
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#1.A
#파일읽기
data = pd.read_csv('au4_2500_train.csv')
test = pd.read_csv('au4_2500_test.csv')

#둘 데이터 합치기
total_data = pd.concat([data,test], axis=0)

#결측값 처리해버리기
new_data = total_data.fillna(method ='ffill')
new_data.info()
```

```python
#가장관련된 coef뽑기위해 sorting
unsortcoef = pd.DataFrame(coef['Class'])
unsortcoef
```

```python
sortcoef = unsortcoef.sort_values(by=['Class'])
sortcoef
```

```python
#테스트데이터넣어서 확인해보기 
final_test = pd.read_csv('au4_2500_test.csv')
final_test = final_test.fillna(method ='ffill')

final_X_test = final_test[final_pick]
final_X_test.info()
#object값 변경
final_X_test.loc[:,'V3'] = le.fit_transform(final_X_test['V3'])
final_X_test.loc[:,'V51'] = le.fit_transform(final_X_test['V51'])
#final_X_test.loc[:,'V100'] = le.fit_transform(final_X_test['V100'])
#정규화하기
final_X_test_normalized = (final_X_test - final_X_test.mean()) / final_X_test.std()


final_Y_test = test_data['Class']
```

```python
final_test = knn.score(final_X_test_normalized,final_Y_test)##test accuracy
final_train = knn.score(final_X_train_normalized, final_Y_train)#training accuracy
display(final_test)
display(final_train)
```

```python
# Decision tree 분류기
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
tree_model.fit(final_X_train_normalized, final_Y_train)
test = tree_model.score(final_X_test_normalized,final_Y_test)
train = tree_model.score(final_X_train_normalized, final_Y_train)
print(test)
print(train)
```

```python
#SVM 
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(final_X_train_normalized, final_Y_train)
test = svm_model.score(final_X_test_normalized,final_Y_test)
train = svm_model.score(final_X_train_normalized, final_Y_train)
print(test)
print(train)
```

```python
#데이터분석2
#최적의 k값 찾기
from sklearn.model_selection import cross_val_score
k_list = list(range(1,30))
k_list = list(filter(lambda x: x%2 == 1, k_list))
knn_error = []

knn_score = []
```

```python
k_list
```

```python
from sklearn.model_selection import train_test_split
for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(final_X_train_normalized, final_Y_train)
    aaaa = knn.score(final_X_test_normalized,final_Y_test)
    knn_score.append(aaaa)
    print(aaaa + k)
max(knn_score)
#k가 9일때 최대가된다 
```

```python
#k가 9일때 0.662 으로 최대값이다 .
knn3 = KNeighborsClassifier(n_neighbors=9)
knn3.fit(final_X_train_normalized, final_Y_train)

final_test = knn3.score(final_X_test_normalized,final_Y_test)##test accuracy
final_train = knn3.score(final_X_train_normalized, final_Y_train)#training accuracy
display(final_test)
display(final_train)
```

```python
#svm 최적화 
from sklearn import svm
import numpy as np
#C가 크면 허용범위가 작아짐 즉 엄격해짐 
#gamma 바꿔보기
for i in range(1,5):
    clf = svm.SVC(gamma=i, C=1)
    clf.fit(final_X_train_normalized, final_Y_train)
    clf.predict(final_X_test_normalized)
    a = clf.score(final_X_test_normalized,final_Y_test)
    print( clf.score(final_X_test_normalized,final_Y_test))
    #gamma가 2일때 최대값을 가짐 
```

```python
#c 최적값을 위해 test
c_list = list(range(1,30))
c_list = list(filter(lambda x: x%2 == 1, c_list))

svm_error = []
svm_score = []
c_list
```

```python
for c in c_list:
    svm_cv2 = svm.SVC(gamma=2, C=c)
    svm_cv2.fit(final_X_train_normalized, final_Y_train)
    svm_score = svm_cv2.score(final_X_test_normalized,final_Y_test)
    print(svm_score)
# c가 3이상이면 0.66으로 나온다 
```

```python
#confusion matrix
#9knn일때 매트릭스 
from sklearn.metrics import confusion_matrix, classification_report
Y_pred = knn3.predict(final_X_test_normalized)
confusion = confusion_matrix(final_Y_test, Y_pred)
print(confusion)
print(classification_report(final_Y_test, Y_pred))
```

```python
#confusion matrix
#gamma가 2 c가 3이상의 svm의 매트릭스 
svm_cv3 = svm.SVC(gamma=2, C=3)
svm_cv3.fit(final_X_train_normalized, final_Y_train)
Y_pred2 = svm_cv3.predict(final_X_test_normalized)
confusion2 = confusion_matrix(final_Y_test, Y_pred2)
print(confusion2)
print(classification_report(final_Y_test, Y_pred2))
```
